```python
import numpy as np
import pandas as pd
```

```python
nfl_data = pd.read_csv("data.csv")
```

```python
#first five rows of the dataset
nfl_data.head()
```

```python
#generating random value everytime the same random values are generated, set seed for reproducibility, used for debugging
np.random.seed(0)
```

```python
missing_values_count = nfl_data.isnull().sum()
missing_values_count
```

```python
missing_values_count[0:10]
#says missing values in the first 10 columns
```

```python
nfl_data.dropna()
#drops all NA values, we find the entire data is removed and rows are zero. so we cannot drop NA values, so we can drop columns with at least one NA Values 
```

```python
col_with_na_dropped = nfl_data.dropna(axis=1)
#when we drop NA values in columns there is 50% of data removed, so this is also not the right thing, so te only possible thin is one s to fill the missed data
```

```python
col_with_na_dropped.head()
```

```python
print("columns in original data : %d \n" % nfl_data.shape[1])
print("columns with NA dropped : %d" % col_with_na_dropped.shape[1])
```

```python
subset_nfl_data = nfl_data.loc[:,'EPA':"Season"].head()
```

```python
subset_nfl_data
```

```python
#fills all NA valus with 0
subset_nfl_data.fillna(0)
```

```python
#replace all NA values that come directly after it in te same column using bfill or back fill
#then replace the remaining NA values with zero

subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```

```python
import numpy as np
import pandas as pd
data = pd.read_csv('Building_Permits.csv',error_bad_lines=False, engine="python")
```

```python
np.random.seed(0)
```

```python
missing_values_count1 = data.isnull().sum()
#isnull says if the value is null or not, so sum of isnull gives the count of null true values
```

```python
total_number = np.product(data.shape)
```

```python
data.shape
```

```python
total_missing = missing_values_count1.sum()
```

```python
print('percent: ',(total_missing / total_number)*100)
```
